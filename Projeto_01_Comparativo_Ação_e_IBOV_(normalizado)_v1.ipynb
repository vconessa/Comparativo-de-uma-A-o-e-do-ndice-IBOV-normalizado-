```python
ticker = str(input('Ação: '))
data_i = str(input('Data início (aaaa-mm-dd): '))
data_f = str(input('Data fim (aaaa-mm-dd):  '))
```

```python
df_ibov = yf.download('^BVSP',start=data_i,end=data_f, actions=True).reset_index()
df_ibov.head()
```

```python
type(df_ibov)
```

```python
df_ibov.dtypes
```

```python
plt.figure(figsize=(15,8))
plt.plot(df_ibov['Adj Close'])
plt.title(('IBOV - '+(data_i)+'<>'+(data_f)), fontsize=15, loc='Center')
plt.show()
```

```python
df_ticker = yf.download((ticker),start=data_i,end=data_f,actions=True).reset_index()
df_ticker
```

```python
plt.figure(figsize=(15,8))
plt.plot(df_ticker['Adj Close'])
plt.title((ticker)+' - ' +(data_i)+'<>'+(data_f), fontsize=15, loc='Center')
plt.show()
```

```python
df_comparativo = pd.DataFrame((df_ibov['Adj Close'], df_ticker['Adj Close'])).T
df_comparativo.head()
```

```python
# Alterando nome das colunas
df_comparativo.columns = ['Close IBOV', ((ticker))]
df_comparativo.head()
```

```python
# Normalização usando escalonamento de recurso min-max 
df_comparativo_min_max = df_comparativo.copy()
for column in df_comparativo_min_max.columns:
  df_comparativo_min_max[column] = (df_comparativo_min_max[column] - df_comparativo_min_max[column].min()) / (df_comparativo_min_max[column].max() - df_comparativo_min_max[column].min())
df_comparativo_min_max.head()
```

# Gráfico sem normalização 

```python
plt.figure(figsize=(15,8))
plt.plot(df_ticker
         ['Adj Close'])
plt.plot(df_ticker['Dividends'])
plt.title('COMPARATIVO * '+' '+(ticker)+' - IBOV  '+(data_i)+'<>'+(data_f), fontsize=15, loc='Center')
plt.show()
```

## Gráfico Normalizado

```python
plt.figure(figsize=(15,8))
plt.plot(df_comparativo_min_max['Close IBOV'])
plt.plot(df_comparativo_min_max[(ticker)])
plt.title('Comparação Normalizada (Escalonamento MIN/MAX) - '+(data_i)+'<>'+(data_f))
plt.legend(['IBOV',(ticker)])
plt.show()
```
